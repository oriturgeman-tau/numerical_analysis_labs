### Lagrange Polynomials

The Lagrange method is an interpolation method—we have data points, but we don't have a function. 

We want to construct a function that passes through all of our known points, so that if we want to calculate the $y$ value of an $x$ we didn't measure, we will have a formula to plug it into. 

It is excellent for finding a function given a small number of points with highly varying intervals (unlike Newton's method, which requires equally spaced intervals). For a large number of points, it becomes inaccurate, and it is better to use splines (which we will cover in the upcoming lessons).

### Gauss-Seidel

The Gauss-Seidel method is a method for solving systems of linear equations. 

It works best for matrices with a dominant diagonal, and is guaranteed to converge for symmetric and positive-definite matrices. Therefore, we will always choose to use it in such cases. It is much less efficient for massive, dense matrices where all values are known in advance—but it is excellent for massive sparse matrices (with a lot of zeros) or for large matrices to which we might want to add values later on.
