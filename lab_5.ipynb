# Splines: Local Interpolation

When we have a dataset (such as outdoor temperature measurements taken at different hours), we prefer to fit a continuous curve through the data points, as it is mathematically easier to handle. Intuitively, our first thought might be to use a high-degree polynomial that passes through all of them.

The problem, however, is that high-degree polynomial curves tend to "go wild" near the boundaries, generating severe oscillations and massive waves. This makes our approximation physically unrealistic and, above all, useless.

**The purpose of splines is to solve this problem locally.** We divide the domain into smaller intervals and fit a low-degree polynomial (such as 1st, 2nd, or 3rd order) between each pair of adjacent points. We can then construct these segments so that they connect smoothly and continuously.

A few examples we covered in the course: *Linear Spline:* Useful when the data points are very dense, Fast. *Hermite Interpolation:* Useful when we have predefined physical data regarding the slopes.

What distinguishes the different types is the number of conditions that must be satisfied for each (e.g., continuity of the first derivative, continuity of the second derivative, etc.).

In today's TA, we will focus on the most widely used method — the **Cubic Spline**.

## Today's topic: Cubic spline


**Why is this a good method?**
- The algorithm works great even if the datapoints are not in equal distances from each other. 
- It achieves continuity in both "velocity" (1st derivative) and "acceleration" (2nd derivative), making it the best approach for describing phenomena observed in experiments or in nature in general.
- If we only know specific data points but not their rate of change, it can perfectly predict what that rate of change should be. (On the other hand, if we already possess data regarding the slope, we would prefer to use Hermite interpolation).
